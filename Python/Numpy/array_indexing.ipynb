```python
import numpy as np
```

**NUMPY INDEXING AND SLICING OF VECTORS**

```python
arr = np.arange(11, 21)
arr
```

```python
arr_slice = arr[2:5] # (start_idx, end_idx) -> (i0,j0)
arr_slice
# i0, i1, i2, i3, ......, j3, j2, j1 -> means : include first_index and (last_index - 1)
```

```python
arr_slice = arr[3::2] # print elements from the start_idx to the last element with k gaps
arr_slice
# (start_idx :: k)
```

```python
arr_slice = arr[:5] # or (0:5)
arr_slice
```

```python
arr_slice = arr[3:] # or (start_idx : last_element)
arr_slice
```

**NUMPY INDEXING AND SLICING OF MATRIX**

```python
arr = np.arange(1,31).reshape((6,5))
arr
```

```python
arr[0] # print index-0 row
```

```python
arr[4] # print index-4 row 
```

```python
arr[:,0] # print index-0 column
```

```python
arr[:,3] # pritn index-3 column
```

```python
arr[3:, 3:] # portion of matrix using slicing
```
