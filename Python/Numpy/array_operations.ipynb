**Deep and Shallow copy**

```python
original_array = np.array([1, 2, 3, 4, 5])
shallow_copy = original_array.view()
# Modify the shallow copy
shallow_copy[0] = 99
print(f"Original array: {original_array}") # Output: [99, 2, 3, 4, 5]
print(f"Shallow copy: {shallow_copy}")     # Output: [99, 2, 3, 4, 5]
```

```python
original_array = np.array([1, 2, 3, 4, 5])
deep_copy = original_array.copy()
# Modify the deep copy
deep_copy[0] = 99
print(f"Original array: {original_array}") # Output: [1, 2, 3, 4, 5]
print(f"Deep copy: {deep_copy}")     # Output: [99, 2, 3, 4, 5]
```
