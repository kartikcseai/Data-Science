**Array Generation Function** 

```python
arr = np.arange(1,11) # make an array of elements from 1 to 10 (1, n+1) 
arr
arr2 = np.arange(1,11,2) # 2 ke gap par elements honge 
arr2
```

```python
arr = np.zeros(5) # array with all elements as "0"
arr
arr2 = np.zeros((4,4)) # n x n matrix with all elements as "0"
arr2
```

```python
arr = np.ones(5) # array with all elements as "1"
arr
arr2 = np.ones((6,6)) # n x n matrix with all elements as "1"
arr2
```

```python
# for linear spacing between elements
arr = np.linspace(0,10,2) # (start_point, end_point, number_of_elements) -> yeh end point tak numbers dega means end_point is included
arr
arr2 = np.linspace(0,10,100)
arr2
```
